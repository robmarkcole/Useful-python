```python
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid")

# Load one of the data sets that come with seaborn
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='scatter');
```

```python
sns.lmplot("total_bill", "tip", tips, col="smoker");
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()

df['x'] = random.sample(range(0, 25), 25)
df['y'] = random.sample(range(0, 25), 25)
df['z'] = random.sample(range(0, 25), 25)
df.head()
```

```python
# sns.lmplot(x = 'x', y = 'y', data = df, fit_reg=False);  #scatter plot
```

```python
df_YZ = pd.melt(df, id_vars=['x'], value_vars=['y', 'z'])
df_YZ
```

Dataframe must be in a particular format for seaborn plots, seee http://stackoverflow.com/questions/29712334/plot-multiple-columns-on-same-graph-seaborn

http://pandas.pydata.org/pandas-docs/stable/generated/pandas.melt.html

To use these features, your data has to be in a Pandas DataFrame and it must take the form of what Hadley Whickam calls tidy data. In brief, that means your dataframe should be structured such that each column is a variable and each row is an observation.

```python
sns.lmplot("x", "value", df_YZ, col="variable");
```

```python
g = sns.factorplot(x="x", y="value", hue="variable", data=df_YZ, kind="violin")
```

```python
sns.factorplot(x="x", y="value", hue="variable", data=df_YZ)  # , kind="bar"
```
